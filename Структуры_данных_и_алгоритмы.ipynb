```python
def encrypt(string):

    alf = ('а','б','в','г','д','е','ё','ж','з','и','й','к','л','м','н','о','п','р','с','т','у','ф','х','ц','ч','ш','щ','ъ','ы','ь','э','ю','я')
    new = []
    for i in string:
        if i not in alf:
            new.append(i)
        else:
            if  i == 'я':
                new.append('а')
            else:
                new.append(alf[alf.index(i)+1])
    return "*".join(new)
word = input('Введите слово для шифрования')
print(encrypt(word))
```

## Связный список

```python
class Node:
    def __init__(self, cargo=None, next=None, back=None):
        self.cargo = cargo
        self.next  = next
        self.back = back

    def __str__(self):
        return str(self.cargo)
    
```

```python
node = Node("test")

print(node)
```

```python
node1 = Node(1)
node2 = Node(2)
node3 = Node(3)
```

```python
node1.next = node2
node2.next = node3
node2.back = node1
```

# Стек

```python
class Stack :
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        return self.items.pop()

    def is_empty(self):
        return (self.items == [])
```

```python
s = Stack()
s.push(54)
s.push(45)
s.push("+")
while not s.is_empty():
    print(s.pop())
```

```python
from collections import deque
q = deque()
 
q.append('eat')
q.append('sleep')
q.append('code')
 
print(q)
# deque(['eat', 'sleep', 'code'])
 
print(q.popleft()) # 'eat'
print(q.popleft()) # 'sleep'
print(q.popleft()) # 'code'
```

```python
mid = len(A) // 2
low = 0
high = len(A) - 1
s=0
while low <= high:
    if K<A[0]:
        s=0
        break
    elif K>A[len(A)-1]:
        s = len(A)-1
        break
    elif A[mid]==K:
        s=mid-1
        break
    elif A[mid] < K < A[mid + 1]:
        s = mid
        break
    elif K > A[mid]:
        low = mid + 1
    else:
        high = mid - 1
    mid = (low + high) // 2
```
